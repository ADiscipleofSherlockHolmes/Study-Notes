### 简单最小二乘回归

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

#### 导入数据 data.csv

```python
points = pd.read_csv("data.csv",header=None)
```

```python
points
```

```python
x = np.array(points[0])
y = np.array(points[1])
plt.scatter(x,y,color='c')
plt.show()
```

```python
# w，b
x = np.array(points[0])
x_bar = np.average(x)
y = np.array(points[1])
length = points.shape[0]
sum_yx = (y*(x-x_bar)).sum()
sum_x2 = (x*x).sum()
sum_1m = length*(x_bar**2)
w = sum_yx/(sum_x2-sum_1m)
b = ((y-w*x).sum())/length
```

```python
y_pred = w*x+b
```

```python
plt.scatter(x,y,color='c')
plt.plot(x,y_pred,color='b')
plt.show()
```

```python
cost = (((y-w*x-b)**2).sum())/length
print("cost:",cost)
```
